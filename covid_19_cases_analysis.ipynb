```
def plotCountryData(country,data,name):
  dataIndia = data[data['countryterritoryCode']==country]
  from datetime import datetime
  dataIndia['dateRep'] = dataIndia['dateRep'].apply(lambda x: datetime.strptime(x,"%d/%m/%Y"))
  dataIndia.sort_values(by=['dateRep'],inplace=True)
  dataIndia.reset_index(inplace=True)
  acutalData = dataIndia
  fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,6))
  dataIndia.plot('dateRep','cases',color='white',ax=axes[0])
  # String giving your date format as "MM-DD" e.g. "04-29"
  date_fmt = '%m-%d'
  date_formatter = mdate.DateFormatter(date_fmt)
  axes[0].xaxis.set_major_formatter(date_formatter)
  fig.autofmt_xdate() # Sets your dates slightly diagonal to fit better.
  #set background color
  axes[0].set_facecolor("green")
  #set x label
  font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
  axes[0].set_xlabel("Dates Wise",fontdict=font)
  axes[0].set_ylabel("coronavirus",fontdict=font)
  axes[0].text(0.5*(left+right), 0.5*(bottom+top), name,
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=20, color='white',
        transform=axes[0].transAxes)
  dataIndia.plot('dateRep','deaths',color='white',ax=axes[1])
  ax = plt.gca() # Gets the current axes
  # String giving your date format as "MM-DD" e.g. "04-29"
  date_fmt = '%m-%d'
  date_formatter = mdate.DateFormatter(date_fmt)
  ax.xaxis.set_major_formatter(date_formatter)
  fig.autofmt_xdate() # Sets your dates slightly diagonal to fit better.
  #set background color
  ax.set_facecolor("green")
  #set x label
  font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
  axes[1].set_xlabel("Dates Wise",fontdict=font)
  axes[1].set_ylabel("Death Count",fontdict=font)
  axes[1].text(0.5*(left+right), 0.5*(bottom+top), name,
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=20, color='white',
        transform=axes[1].transAxes)

  fig.tight_layout()

```

```
plotCountryData('USA',data,"America")
```

```
def plotTwoCountriesData(country1,country2,data,name):
  dataCountry1 = data[data['countryterritoryCode']==country1]
  dataCountry2 = data[data['countryterritoryCode']==country2]
  
  from datetime import datetime

  dataCountry1['dateRep'] = dataCountry1['dateRep'].apply(lambda x: datetime.strptime(x,"%d/%m/%Y"))
  dataCountry1.sort_values(by=['dateRep'],inplace=True)
  dataCountry1.reset_index(inplace=True)

  dataCountry2['dateRep'] = dataCountry2['dateRep'].apply(lambda x: datetime.strptime(x,"%d/%m/%Y"))
  dataCountry2.sort_values(by=['dateRep'],inplace=True)
  dataCountry2.reset_index(inplace=True)

  fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,6))

  dataCountry1.plot('dateRep','cases',color='red',ax=axes[0],label=country1)
  dataCountry2.plot('dateRep','cases',color='white',ax=axes[0],label=country2)

  # String giving your date format as "MM-DD" e.g. "04-29"
  date_fmt = '%m-%d'
  date_formatter = mdate.DateFormatter(date_fmt)
  axes[0].xaxis.set_major_formatter(date_formatter)
  fig.autofmt_xdate() # Sets your dates slightly diagonal to fit better.
  #set background color
  axes[0].set_facecolor("green")
  
  #set x label
  font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
  axes[0].set_xlabel("Dates Wise",fontdict=font)
  axes[0].set_ylabel("coronavirus cases",fontdict=font)
  axes[0].text(0.5*(left+right), 0.5*(bottom+top), name,
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=20, color='white',
        transform=axes[0].transAxes)
  dataCountry1.plot('dateRep','deaths',color='red',ax=axes[1],label=country1)
  dataCountry2.plot('dateRep','deaths',color='white',ax=axes[1],label=country2)

  ax = plt.gca() # Gets the current axes
  # String giving your date format as "MM-DD" e.g. "04-29"
  date_fmt = '%m-%d'
  date_formatter = mdate.DateFormatter(date_fmt)
  ax.xaxis.set_major_formatter(date_formatter)
  fig.autofmt_xdate() # Sets your dates slightly diagonal to fit better.
  #set background color
  ax.set_facecolor("green")
  #set x label
  font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
  axes[1].set_xlabel("Dates Wise",fontdict=font)
  axes[1].set_ylabel("Death Count",fontdict=font)
  axes[1].text(0.5*(left+right), 0.5*(bottom+top), name,
        horizontalalignment='center',
        verticalalignment='center',
        fontsize=20, color='white',
        transform=axes[1].transAxes)

  fig.tight_layout()

```

```
plotTwoCountriesData('USA','IND',data,'USA/IND')
```

```
plotTwoCountriesData('USA','CHN',data,'USA/CHN')
```
